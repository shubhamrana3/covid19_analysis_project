```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv("/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
# lets clean our dataset, we don't need latitudes,longitudes here.
dataset.shape
```

```python
# let delete the useless columns
df = dataset.drop(["Lat","Long"],axis=1,inplace=True)
dataset.head()
```

```python
# aggregate the rows by the country.
corona_dataset_aggregated=dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# calculating the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
#now we will find out the maximum infection rate for all the countries and store it in our dataset
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]
for c in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rates"]=max_infection_rates
corona_dataset_aggregated
```

```python
#now we will create a new dataframe with only the columns which we need,we are not intrested in all the dates which we have.
# creating a new dataframe
corona_data=pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])
corona_data
# now we have different countries and their maximum infection rates
```

```python
# now we will import world happiness report csv file and select the columns that we need for analysis
#importing second dataset
happiness_report=pd.read_csv("/worldwide_happiness_report.csv")
happiness_report
```

```python
# dropping the useless columns.
useless_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols, axis=1, inplace = True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

```python
#now we will join the datasets,but before joining we will look at the shape of the datasets.
corona_data.shape
```

```python
happiness_report.shape
```

```python
data=corona_data.join(happiness_report,how="inner")
data
#Inner Join:- it returns a dataframe with only those rows that have common characteristics.
```

```python
#now we will go ahead and create a corelation matrix to see the corelation between these factors
data.corr()
```
